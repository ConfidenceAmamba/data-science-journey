```python
#creating a function WITH CONDITIONALS
def mean(value):  #syntax
    if type(value) == dict:
        the_mean=sum(value.values())/len(value)
    else:
        the_mean=sum(value)/len(value)
            
    return the_mean                  #SYNTAX
print(mean([4,6,7,8,9,9,67,67,8]))
print(mean({"success":  6.7,"Jason": 5.8,"Flora": 9.0,"jjj"  :5.9,"roy" :4.5}))
```
